```python
import json
import os
import boto3

# Environment variables
TEMPLATE_NAME = "delivery_status"
TEMPLATE_LANGUAGE = "es_CL"
ORIGINATION_PHONE_NUMBER_ID = "XXX"
META_API_VERSION = "v23.0"
phone_number = "+XXX"

client = boto3.client("socialmessaging")
```

```python
template_components = [{
    "type": "body",
    "parameters": [
        {"type": "text", "text": "value1"},
        {"type": "text", "text": "value2"},
        {"type": "text", "text": "value3"},
        {"type": "text", "text": "value4"}
    ]
}]


message_payload = {
    "messaging_product": "whatsapp",
    "to": phone_number,
    "recipient_type": "individual",
    "type": "template",
    "template": {
        "name": TEMPLATE_NAME,
        "language": {"code": TEMPLATE_LANGUAGE},
    }
}

# Add components only if provided
if template_components:
    message_payload["template"]["components"] = template_components

print(f"Sending template '{TEMPLATE_NAME}' to {phone_number}")
print(message_payload)

response = client.send_whatsapp_message(
    originationPhoneNumberId=ORIGINATION_PHONE_NUMBER_ID,
    message=bytes(json.dumps(message_payload), "utf-8"),
    metaApiVersion=META_API_VERSION,
)

message_id = response.get("messageId", "")
print(f"Message sent successfully. messageId: {message_id}")
```
